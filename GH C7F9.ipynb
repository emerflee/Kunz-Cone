```python
#previously,
# for face in C7lat:
#     if face.dim() == 5:
#         print(face)
#         print(face.rays())
#         print(face.ambient_Hrepresentation())
#then took the rays for the first 9-ray facet
C7F9 = Polyhedron(rays = [(5,3,1,6,4,2),(8,2,10,4,5,6),(8,2,3,4,5,6),(12,3,8,6,4,9),(5,3,8,6,4,2),(4,1,5,2,6,3),(6,5,4,3,2,1),(9,4,6,8,3,5),(10,6,2,5,8,4)])
#take the cross section of that
P7F9 = C7F9.intersection(Polyhedron(eqns = [[-1,1,1,1,1,1,1]]))
```

```python
#P7F9.vertices() to get the vertex list
VA = (5/21, 1/7, 1/21, 2/7, 4/21, 2/21)
VB = (4/21, 1/21, 5/21, 2/21, 2/7, 1/7)
VC = (2/7, 5/21, 4/21, 1/7, 2/21, 1/21)
VD = (2/7, 6/35, 2/35, 1/7, 8/35, 4/35)
VE = (2/7, 1/14, 3/28, 1/7, 5/28, 3/14)
VF = (5/28, 3/28, 2/7, 3/14, 1/7, 1/14)
VG = (2/7, 1/14, 4/21, 1/7, 2/21, 3/14)
VH = (8/35, 2/35, 2/7, 4/35, 1/7, 6/35)
VI = (9/35, 4/35, 6/35, 8/35, 3/35, 1/7)
```

```python
#the manual placements for the vertices of the Schlegel diagram of P7F9. 
V1 = (0,0,0)
V2 = (4,0,4)
V3 = (2,3,3)
V4 = (1,1,3.5)
V5 = (0,0,4)
V6 = (4,4,0)
V7 = (0,4,4)
V8 = (4,4,4)
V9 = (0,4,0)
```

```python
vertpoints = [V1, V2, V3, V4, V5, V6, V7, V8, V9] #list of points
vertnames = {V1: 'A1', V2: 'B2', V3: 'C3', V4: 'D4', V5: 'E5', V6: 'F6', V7: 'G7', V8: 'H8', V9: 'I9'}
#tells which actual vertex (letters) get placed where (numbers)
E = {VA:V1, VB:V2, VC:V3, VD:V4, VE:V5, VF:V6, VG:V7, VH:V8, VI:V9} 
```

```python
#tell sage how to build the schlegel diagram
def schlegeldiagram(C):
    Pts=points(vertpoints, size=0, frame=False)
    for pt in vertpoints:
        Pts = Pts + point3d(pt, size=70, color = "blue")
        Pts = Pts+ text3d(vertnames[pt], pt, fontsize = 30)
    for F in C.faces(1):
        EE=F.ambient_V_indices()
        Pts=Pts + line([vertpoints[EE[0]], vertpoints[EE[1]]], color = "blue")
    return Pts
```

```python
#actually look at the diagram! nice!
schlegeldiagram(P7F9)
```
